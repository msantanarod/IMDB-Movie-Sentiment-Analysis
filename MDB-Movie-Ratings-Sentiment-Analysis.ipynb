## Text data cleaning

```python

stop_words = list(stopwords.words('English'))

def preprocess(review):
   
    result = []
    for token in gensim.utils.simple_preprocess(review):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
# apply the function to the dataset:

df['clean_review'] = df['review'].apply(preprocess)
```

```python
# join the words into a string
df['clean_review_joined'] = df['clean_review'].apply(lambda x: " ".join(x))
```

## Sentiment Analysis

Returns results as a named tuple of the form: Sentiment(polarity, subjectivity)

```python
df['sentiment'] = df.clean_review_joined.apply(lambda text: TextBlob(text).sentiment)
df
```

```python
# We want to split 'sentiment' column into two different columns: 'polarity' and 'subjectivity':

df['sentiment'][0][0]
df['polarity'] = df['sentiment'].apply(lambda x: x[0])
df['subjectivity'] = df['sentiment'].apply(lambda x: x[1])
```

```python
df
```

```python
# We create a new column named: 'sentiment_score' based on the 'polarity' column:

conditions = [
    (df['polarity'] <0),
    (df['polarity'] ==0),
    (df['polarity'] > 0)
    ]

# create a list of the values we want to assign for each condition
values = ['negative', 'neutral', 'positive']

# create a new column and use np.select to assign values to it using our lists as arguments
df['sentiment_score'] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

```python
# save the dataset:

df.to_csv("movie_review_sentiment_analysis.csv")

```
